```python
#1 Definimos la función para obtener la gravedad local
def obtener_gravedad_local(latitud: float, longitud: float, api_key: str) -> dict:
    url = 'https://gravity.amentum.io/egm2008/gravity_anomaly'

    # Cambiamos las etiquetas a nombres completos
    params = {
        "latitude": latitud,   # Antes era "lat"
        "longitude": longitud  # Antes era "lon"
    }

    headers = {
        "API-Key": api_key,
        "accept": "application/json"
    }

    import requests
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        return response.json()
    else:
        return {"error": f"Fallo {response.status_code}", "detalle": response.text}
    
```

```python
import pandas as pd
import requests
import io

# 1. La URL debe ser la dirección completa de GitHub (usando raw.githubusercontent.com)
# OJO: He puesto 'proyecto_idividual' porque así aparece en tus capturas anteriores
url = "https://raw.githubusercontent.com/nailaikhenazenni-afk/proyecto_idividual_asteroides/main/asteroids_data.csv"

# 2. Encabezado para simular un navegador (estilo Mozilla que discutimos)
headers = {"User-Agent": "Mozilla/5.0"}

try:
    # 3. Descargamos el contenido
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        # 4. Leemos el CSV (usando ';' como indicas en tu código)
        df = pd.read_csv(io.StringIO(response.text), sep=';')
        
        # Estilo: Usar nombres específicos y configuración limpia
        pd.set_option('display.max_columns', None)
        
        print("¡Dataset cargado con éxito!")
        print(df.head())
    else:
        print(f"Error al descargar: {response.status_code} (Verifica que la URL sea correcta)")

except requests.exceptions.MissingSchema:
    print("Error: La URL no es válida. Asegúrate de que empiece con http:// o https://")
```
