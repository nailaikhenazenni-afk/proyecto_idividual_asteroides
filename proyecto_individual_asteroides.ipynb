```python
#1 Definimos la función para obtener la gravedad local
def obtener_gravedad_local(latitud: float, longitud: float, api_key: str) -> dict:
    url = 'https://gravity.amentum.io/egm2008/gravity_anomaly'

    # Cambiamos las etiquetas a nombres completos
    params = {
        "latitude": latitud,   # Antes era "lat"
        "longitude": longitud  # Antes era "lon"
    }

    headers = {
        "API-Key": api_key,
        "accept": "application/json"
    }

    import requests
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        return response.json()
    else:
        return {"error": f"Fallo {response.status_code}", "detalle": response.text}
    
```

```python
import pandas as pd

try:
    # 1. Cargar el dataset original
    df = pd.read_csv('asteroids_data.csv', sep=',')

    # 2. Lista de columnas a eliminar (incluyendo las técnicas y las que causaban error)
    columnas_a_eliminar = [
        'orbit_id', 'data_arc_days', 'orbit_uncertainty', 'sentry', 'sentry_data', 
        'observations_used', 'jupiter_tisserand', 'epoch', 'eccentricity', 
        'semi_major_axis', 'inclination', 'aphelion_distance', 'perihelion_time', 
        'mean_anomaly', 'neo_id', 'name', 'designation', 'mean_motion', 'equinox', 
        'orbit_class_type', 'orbit_class_desc', 'orbit_class_range', 
        'ascending_node_longitude', 'orbit_determination_date', 
        'perihelion_distance', 'perihelion_argument', 'id', 'api_url'
    ]

    # 3. Aplicar limpieza de columnas y duplicados
    df = df.drop(columns=columnas_a_eliminar, errors='ignore')
    df = df.drop_duplicates()

    # 4. Renombrar short_name a name y limpiar nulos en el nombre
    if 'short_name' in df.columns:
        df = df.rename(columns={'short_name': 'name'})
    
    df = df.dropna(subset=['name'])

    # 5. Gestión de velocidad: Conversión a km/h y eliminación de nulos
    if 'velocidad_km_s' in df.columns:
        # Eliminamos filas donde la velocidad sea nula antes de calcular
        df = df.dropna(subset=['velocidad_km_s'])
        # Creamos la columna en km/h
        df['velocidad_km_h'] = df['velocidad_km_s'] * 3600

    # 6. Guardar el archivo definitivo
    df.to_csv('dataset_asteroides_final.csv', index=False)

    print(f"✅ Archivo 'dataset_asteroides_final.csv' creado con éxito.")
    print(f"📏 Shape final: {df.shape}")
    print(f"📊 Columnas resultantes: {list(df.columns)}")

except Exception as e:
    print(f"❌ Error al procesar el dataset: {e}")
```

```python
import pandas as pd
import ast
import mysql.connector

try:
    # 1. Cargar el dataset original
    df = pd.read_csv('asteroids_data.csv', sep=',')

    # --- EXTRACCIÓN DE DATOS (Basado en image_865674.png) ---
    def extract_nasa_data(row):
        try:
            # Convertimos el string de la celda en una lista real de Python
            data_list = ast.literal_eval(row)
            if data_list and len(data_list) > 0:
                first_approach = data_list[0]
                date = first_approach.get('close_approach_date')
                # Extraemos velocidad en km/s para el cálculo
                vel_s = first_approach.get('relative_velocity', {}).get('kilometers_per_second', 0)
                return pd.Series([date, float(vel_s)])
        except:
            pass
        return pd.Series([None, 0.0])

    print("🔍 Extracting date and velocity from nested data...")
    df[['approach_date', 'velocity_km_s']] = df['close_approach_data'].apply(extract_nasa_data)

    # 2. Renombrar short_name a 'name' (según tus cambios previos)
    if 'short_name' in df.columns:
        df = df.rename(columns={'short_name': 'name'})

    # 3. Cálculo de velocidad en km/h y limpieza
    df['velocity_km_h'] = df['velocity_km_s'] * 3600
    
    # Seleccionamos solo las columnas que pediste en inglés
    df_final = df[['name', 'approach_date', 'velocity_km_h']].copy()
    
    # Limpieza: eliminamos filas con nulos y ordenamos de forma descendente por velocidad
    df_final = df_final.dropna().sort_values(by='velocity_km_h', ascending=False)

    # 4. Guardar el CSV limpio (English version)
    df_final.to_csv('dataset_asteroides_final.csv', index=False)
    print("✅ 'dataset_asteroides_final.csv' created successfully.")

    # --- MYSQL INTEGRATION ---
    config = {
        'user': 'root',
        'password': 'kP92mXv7R4wQ', 
        'host': '127.0.0.1',
        'database': 'asteroides_listo_para_sql'
    }

    conn = mysql.connector.connect(**config)
    cursor = conn.cursor()

    # Recreamos la tabla con los nombres en inglés
    cursor.execute("DROP TABLE IF EXISTS `asteroides_ranking`")
    cursor.execute("""
        CREATE TABLE `asteroides_ranking` (
            `name` VARCHAR(250) PRIMARY KEY,
            `approach_date` DATE,
            `velocity_km_h` FLOAT
        )
    """)

    # Inserción masiva de los datos ya ordenados
    sql_insert = "INSERT INTO `asteroides_ranking` (name, approach_date, velocity_km_h) VALUES (%s, %s, %s)"
    
    # Convertimos el DataFrame a lista de tuplas para la carga
    data_to_insert = [tuple(x) for x in df_final.values]
    cursor.executemany(sql_insert, data_to_insert)
    
    conn.commit()
    print(f"✅ MySQL table updated: {cursor.rowcount} rows inserted in descending order.")

    # 5. Consulta para verificar el Top 10 para tu video
    print("\n🎬 TOP 10 FOR THE VIDEO:")
    print("-" * 50)
    for row in data_to_insert[:10]:
        print(f"Asteroid: {row[0]:<20} | Speed: {row[2]:>10.2f} km/h | Date: {row[1]}")

except Exception as e:
    print(f"❌ Error: {e}")
finally:
    if 'conn' in locals() and conn.is_connected():
        cursor.close()
        conn.close()
```
