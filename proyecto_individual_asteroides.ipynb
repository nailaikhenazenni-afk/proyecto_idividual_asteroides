```python
# Definimos la función para obtener la gravedad local
def obtener_gravedad_local(latitud: float, longitud: float, api_key: str) -> dict:
    url = 'https://gravity.amentum.io/egm2008/gravity_anomaly'

    # Cambiamos las etiquetas a nombres completos
    params = {
        "latitude": latitud,   # Antes era "lat"
        "longitude": longitud  # Antes era "lon"
    }

    headers = {
        "API-Key": api_key,
        "accept": "application/json"
    }

    import requests
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        return response.json()
    else:
        return {"error": f"Fallo {response.status_code}", "detalle": response.text}
    
```

 - Segunda etapa:
  Creación de las tablas:
  Ahora sí, con el dataset limpio y listo, lo cargamos en MySQL y sacamos las tablas que vamos a utilizar para nuestro análisis. vamos a realizar una para obtener el ranking de los 5 asteroides más rápidos que se acercaron a la Tierra, ordenados de mayor a menor velocidad, con su fecha de acercamiento y su nombre. Luego vamos a realizar otra,de donde sacamros un ranking de los 5 asteroides más masívos, sera compusta por su diametro medio que obtendremos, calculando la media de su diametro máximo y mínimo, de su magnetude absoluta y de su nombre. 
  En un segundo tiempo recopilaremos los valores  de anomalia gravitacional sacados de la API con la localizaciones que elgimos con sus coordenadas. Ordenaremos estos valores de manera descendiente.

```python
# Hacemos la tabla de los asteroides más rápidos usando los datos de MySQL
import pandas as pd
import ast
import mysql.connector

try:
    # 1. Cargar el archivo actual
    df = pd.read_csv('dataset_asteroides_final.csv')

    # 2. Función para extraer fecha y velocidad
    def extract_details(row):
        try:
            data = ast.literal_eval(row)
            if isinstance(data, list) and len(data) > 0:
                first_entry = data[0]
                date = first_entry.get('close_approach_date')
                # Velocidad en km/s
                vel_s = first_entry.get('relative_velocity', {}).get('kilometers_per_second', 0)
                return pd.Series([date, float(vel_s)])
        except:
            pass
        return pd.Series([None, 0.0])

    # Extraemos los datos necesarios
    df[['approach_date', 'velocity_km_s']] = df['close_approach_data'].apply(extract_details)

    # 3. Calcular velocidad en km/h
    df['velocity_km_h'] = df['velocity_km_s'] * 3600
    
    # 4. ORDEN DESCENTE: Filtramos, limpiamos y ordenamos de mayor a menor
    df_sql = df[['name', 'approach_date', 'velocity_km_h']].dropna().copy()
    
    # Ordenamos explícitamente aquí por velocidad descendente
    df_sql = df_sql.sort_values(by='velocity_km_h', ascending=False)

    # --- CONFIGURACIÓN MYSQL ---
    config = {
        'user': 'root',
        'password': 'kP92mXv7R4wQ', 
        'host': '127.0.0.1',
        'database': 'asteroides_listo_para_sql'
    }

    conn = mysql.connector.connect(**config)
    cursor = conn.cursor()

    # 5. Recrear tabla
    cursor.execute("DROP TABLE IF EXISTS `asteroides_ranking`")
    cursor.execute("""
        CREATE TABLE `asteroides_ranking` (
            `name` VARCHAR(250) PRIMARY KEY,
            `approach_date` DATE,
            `velocity_km_h` FLOAT
        )
    """)

    # 6. Inserción de los datos (ya vienen ordenados en la lista)
    sql_insert = "INSERT INTO `asteroides_ranking` (name, approach_date, velocity_km_h) VALUES (%s, %s, %s)"
    
    # Convertimos el DataFrame ordenado a una lista de tuplas
    valores = [tuple(x) for x in df_sql.values]
    
    # Insertamos manteniendo el orden del DataFrame
    cursor.executemany(sql_insert, valores)
    conn.commit()

    print(f"✅ Tabla 'asteroides_ranking' cargada en orden descendente.")
    print(f"📊 Total de registros: {cursor.rowcount}")

    # Verificación en consola
    print("\n🔍 TOP 5 MÁS RÁPIDOS:")
    for row in valores[:5]:
        print(f"{row[0]} -> {row[2]:,.2f} km/h")

except Exception as e:
    print(f"❌ Error: {e}")
finally:
    if 'conn' in locals() and conn.is_connected():
        cursor.close()
        conn.close()
```

```python
# Análisis comparativo entre dimensión y velocidad usando los datos de MySQL (usamos views)
import mysql.connector

# configuración de conexión
config: dict = {
    'user': 'root',
    'password': 'kP92mXv7R4wQ', 
    'host': '127.0.0.1',
    'database': 'asteroides_listo_para_sql'
}

def comparar_dimension_vs_velocidad():
    try:
        conn = mysql.connector.connect(**config)
        cursor = conn.cursor()

        # --- 1. ANÁLISIS POR DIMENSIÓN ---
        query_dim = """
        SELECT 
            lider.name,
            lider.diameter_avg_m,
            g.localization,
            g.gravity_value,
            ROUND((lider.diameter_avg_m) / (g.gravity_value / 10), 2) AS impacto
        FROM (
            SELECT name, diameter_avg_m
            FROM dimension_ranking
            ORDER BY diameter_avg_m DESC
            LIMIT 1
        ) AS lider, 
        gravity.gravity AS g
        WHERE g.localization IN ('Monte Everest', 'Fosa de las marianas')
        LIMIT 5;
        """

        # --- 2. ANÁLISIS POR VELOCIDAD ---
        query_vel = """
        SELECT 
            lider.name,
            lider.velocity_km_h,
            g.localization,
            g.gravity_value,
            ROUND(((lider.velocity_km_h / 1000)) / (g.gravity_value / 10), 2) AS impact_velocity
        FROM (
            SELECT name, velocity_km_h
            FROM asteroides_ranking
            ORDER BY velocity_km_h DESC
            LIMIT 1
        ) AS lider, 
        gravity.gravity AS g
        WHERE g.localization IN ('Monte Everest', 'Fosa de las marianas')
        LIMIT 5;
        """

        # Ejecución y visualización
        def imprimir_resultados(titulo, sql):
            cursor.execute(sql)
            resultados = cursor.fetchall()
            print(f"\n" + "="*70)
            print(f" {titulo} ".center(70, "="))
            print(f"{'Objeto':<20} | {'Sede':<20} | {'Valor Atrib.':<12} | {'Impacto':<10}")
            print("-" * 70)
            for r in resultados:
                print(f"{r[0]:<20} | {r[2]:<20} | {r[1]:>12.2f} | {r[4]:>10.2f}")

        imprimir_resultados("IMPACTO BASADO EN DIMENSIÓN", query_dim)
        imprimir_resultados("IMPACTO BASADO EN VELOCIDAD", query_vel)

    except Exception as e:
        print(f"Error en la base de datos: {e}")
    finally:
        if 'conn' in locals() and conn.is_connected():
            cursor.close()
            conn.close()

if __name__ == "__main__":
    comparar_dimension_vs_velocidad()
```

```python
# Realizamos un análisis de impacto unificado combinando dimensión, velocidad y gravedad usando los datos de MySQL
import mysql.connector

config: dict = {
    'user': 'root',
    'password': 'kP92mXv7R4wQ', 
    'host': '127.0.0.1',
    'database': 'asteroides_listo_para_sql'
}

def ejecutar_analisis_unificado() -> None:
    """ejecuta el análisis de impacto unificado y muestra los resultados en consola."""
    try:
        conn = mysql.connector.connect(**config)
        cursor = conn.cursor()

        # sql corregido: une tablas, define el líder por dimensión y ordena por gravedad
        query: str = """
        with lider as (
            select 
                a.name, 
                d.diameter_avg_m,
                a.velocity_km_h
            from asteroides_ranking a
            join dimension_ranking d on a.name = d.name
            order by d.diameter_avg_m desc, a.velocity_km_h desc
            limit 1
        )
        select 
            l.name,
            l.diameter_avg_m,
            l.velocity_km_h,
            g.localization,
            g.gravity_value,
            round((l.diameter_avg_m * (l.velocity_km_h / 1000)) + g.gravity_value, 2) as impacto
        from gravity.gravity g
        join lider l on 1=1
        order by g.gravity_value desc;
        """

        cursor.execute(query)
        dataset: list[tuple] = cursor.fetchall()

        if not dataset:
            print("no se encontraron datos.")
            return

        # datos del asteroide líder (fijos para todas las filas)
        nombre, diametro, velocidad = dataset[0][0], dataset[0][1], dataset[0][2]

        # cabecera del reporte
        print("\n" + "="*95)
        print(f" REPORTE UNIFICADO: {nombre.upper()} ".center(95, "="))
        print(f" Atributos Principales: {diametro:,.2f} m (Diámetro) | {velocidad:,.2f} km/h (Velocidad)")
        print("="*95)
        
        # encabezados de columna
        print(f"{'Localización':<30} | {'Gravedad (mGal)':<18} | {'Índice Impacto':<20}")
        print("-" * 95)

        # iteración y formato de resultados
        for fila in dataset:
            loc, grav, impacto = fila[3], fila[4], fila[5]
            # formateamos con signos para la gravedad y comas para el impacto
            print(f"{loc:<30} | {grav:>18.2f} | {impacto:>20,.2f}")

        print("-" * 95)
        print("nota: ordenado por gravedad descendente (mayor a menor).")

    except Exception as e:
        print(f"error en la base de datos: {e}")
    finally:
        if 'conn' in locals() and conn.is_connected():
            cursor.close()
            conn.close()

if __name__ == "__main__":
    ejecutar_analisis_unificado()
```
